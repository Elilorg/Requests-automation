```python
from tgtg import TgtgClient
import json

#client = TgtgClient()
#client.signup_by_email(email="Dopi21@yopmail.com")



client = TgtgClient(email="Dopi21@yopmail.com")
credentials = client.get_credentials()
print(credentials)
```

```python
items = client.get_items(
    favorites_only=False,
    latitude=48.881, 
    longitude=2.421,
    radius=100,
    page_size=400
)
print(str(items).replace("'", '"'))
```

```python
print(len(items))
```

```python
with open("response.json", "w") as f:
    json.dump(items, f)


liste_items = []
for i in items : 
    item = i["item"]
    prix = item["price_including_taxes"]["minor_units"]/100
    
    
    print(f"""
    Item ID : {item["item_id"]} 
    Prix : {prix}
    Name : {item["name"]}
    Nom magasin : {i["store"]["store_name"]}
    Adresse : {i["store"]["store_location"]["address"]["address_line"]}
    """)
    liste_items.append(
        {
            "id" : item["item_id"], 
            "prix" : prix, 
            "nom" : item["name"], 
            "nom magazin" : i["store"]["store_name"], 
            "adresse" : i["store"]["store_location"]["address"]["address_line"]
        }
    )
print(str(liste_items).replace("'", '"'))
```
